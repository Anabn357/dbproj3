```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into pandas DataFrame
Score_Analysis = pd.read_csv('whatoplay_data.csv')
Score_Analysis.head()
```

```python
# Assign the data to X and y for Play Score Analysis
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

X = Score_Analysis.PlayScore.values.reshape(-1, 1)
y = Score_Analysis.NA_Sales.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
X
```

```python
# Plot the data

### BEGIN SOLUTION
#plt.xlabel("Play Score")
#plt.ylabel("North America Sales(Millions)")
#plt.title("Play Score Analysis")
#plt.figure(figsize=((10,8)))

fig, ax = plt.subplots(figsize=(10,8))
ax.set_title('Play Score Analysis')
ax.scatter(X, y)
ax.set(xlabel='Play Score', ylabel='North America Sales(Millions)',title='Play Score Analysis')
fig.savefig('images/PlayScore.png')







### END SOLUTION
```

```python
# Create the model and fit the model to the data

from sklearn.linear_model import LinearRegression

### BEGIN SOLUTION

model = LinearRegression()

### END SOLUTION
```

```python
# Fit the model to the data. 
# Note: This is the training step where you fit the line to the data.

### BEGIN SOLUTION

model.fit(X, y)

### END SOLUTION
```

```python
# Print the coefficient and the intercept for the model

### BEGIN SOLUTION
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
### END SOLUTION
```

```python
# Note: we have to transform our min and max values 
# so they are in the format: array([[ 1.17]])
# This is the required format for `model.predict()`

x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
# Calculate the y_min and y_max using model.predict and x_min and x_max

### BEGIN SOLUTION
y_min = model.predict(x_min)
y_max = model.predict(x_max)
### END SOLUTION
```

```python
# Plot X and y using plt.scatter
# Plot the model fit line using [x_min[0], x_max[0]], [y_min[0], y_max[0]]

### BEGIN SOLUTION
#plt.scatter(X, y, c='blue')
#plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
#plt.xlabel("Play Score")
#plt.ylabel("Global Sales(Millions)")
#plt.title("(Play Score) Linear Regression Analysis")

#plt.show
### END SOLUTION
fig, ax = plt.subplots(figsize=(10,8))
ax.set_title('Play Score Analysis')
ax.scatter(X, y)
ax.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
ax.set(xlabel='Play Score', ylabel='North America Sales(Millions)',title='(Play Score) Linear Regression Analysis')
fig.savefig('images/PlayScoreRegression.png')

```

## Observation & Conclusions From Machine Learning Analysis:

### The Video Game data had three scores in the form of rating score which were Play Score, Game Score and Critic Score respectively.
### To describe each of them, the Play Score is the score obtained from the Player Ratings on various games, Game Score is the rating given to each game by the developer of the game and the Critic Score is obtained from the critic rating for the games.
### To study the relavance of these scores, we ran some Machine Learning Algorithms on our data (the analysis was done on North America Sales since it is the biggest market in terms of sales), and following are the conclusions that can be drawn:
1. There is a positive corelation between the Video Game Sales and Play Score.
2. An 8-point increase in Play Score seems to have a positive effect on sales of about 250k when starting from a      score of 6.5, but ~1M when starting from 7.7. 
3. There is a positive corelation between the Video Game Sales and Game Score.
4.An 8-point increase in Game Score seems to have a positive effect on sales of about 250k when starting from a score of 6.5, but ~1M when starting from 7.7. 
5. There is a positive corelation between the Video Game Sales and Critic Score.
6.An 8-point increase in Critic Score seems to have a positive effect on sales of about 250k when starting from a score of 6.5, but ~1M when starting from 7.7. 

Note: The results were found to be almost identical for all the three scores.
